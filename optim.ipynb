```python
import ajcs
import sqra
import optim
import matplotlib.pyplot as plt
import numpy as np
import scipy.sparse as sp
```

```python
nx = 5
ny = 2
process = sqra.sqra2d(sqra.doublewell2d(nx=nx, ny=ny), beta=1, phi=1)
process.plot()
```

```python
nt = 4
dt = 1
ts = ([dt] * nt)
```

```python
o=optim.SqraOptim(process, ts, penalty = 0.001, adaptive=True)
```

```python
%time o.optimize(10)
```

```python
o.plot()
```

```python
o.plot_fin_hit_prob()
```
