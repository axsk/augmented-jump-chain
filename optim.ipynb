```python
import ajcs
import sqra
import optim
import matplotlib.pyplot as plt
import numpy as np
import scipy.sparse as sp
```

```python
nx = 5
ny = 3
process = sqra.sqra2d(sqra.doublewell2d, x=(-1.5, 1.5, nx), y=(-1,1,ny), beta=1, phi=1)
process.plot()
```

```python
nt = 5
dt = 1
ts = ([dt] * nt)
```

```python
o=optim.SqraOptim(process, ts, penalty = 0.001, adaptive=True)
```

```python
%time o.optimize(1000)
```

```python
o.plot()
```

```python
plt.imshow(o.sqra.u + np.reshape(o.x, np.shape(o.sqra.u)))
plt.colorbar()
```

```python
plt.imshow(o.j.finite_time_hitting_probs())
plt.colorbar()
plt.title("fin. time hitting prob")
plt.ylabel("start")
plt.xlabel("hit");
```
