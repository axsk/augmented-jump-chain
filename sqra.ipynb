```python
import sqra
import ajcs
```

```python
nx = 5
ny = 2
process = sqra.sqra2d(sqra.doublewell2d, x=(-1.5, 1.5, nx), y=(-1,1,ny), beta=1, phi=1)
process.plot()
```

```python
nt = 5
dt = 1
Qs = [process.Q] * nt
ts = ([dt] * nt)
j = ajcs.AJCS(Qs, ts)
```

```python
q = j.finite_time_hitting_prob((nx-1) + 2)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(q)
```
